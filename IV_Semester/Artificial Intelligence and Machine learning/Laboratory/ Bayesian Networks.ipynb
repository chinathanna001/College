```python
from pgmpy.models import BayesianNetwork
from pgmpy.factors.discrete import TabularCPD

model = BayesianNetwork([('A', 'B'), ('C', 'B'), ('B', 'D'), ('B', 'E')])

cpd_a = TabularCPD(variable='A', variable_card=2, values=[[0.3], [0.7]])
cpd_c = TabularCPD(variable='C', variable_card=2, values=[[0.6], [0.4]])
cpd_b = TabularCPD(variable='B', variable_card=2,
                   values=[[0.9, 0.7, 0.8, 0.1], [0.1, 0.3, 0.2, 0.9]],
                   evidence=['A', 'C'], evidence_card=[2, 2])
cpd_d = TabularCPD(variable='D', variable_card=2,
                   values=[[0.95, 0.25], [0.05, 0.75]],
                   evidence=['B'], evidence_card=[2])
cpd_e = TabularCPD(variable='E', variable_card=2,
                   values=[[0.9, 0.2], [0.1, 0.8]],
                   evidence=['B'], evidence_card=[2])

model.add_cpds(cpd_a, cpd_c, cpd_b, cpd_d, cpd_e)

print(model.check_model())
```
