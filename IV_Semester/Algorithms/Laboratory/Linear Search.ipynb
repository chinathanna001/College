```python
import time
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
time_taken_linear = []
no_of_elem = []
keys = []
```

```python
def linear_search(a, k):
    for i in range(len(a)):
        if a[i] == k:
            return i
    return -1
```

```python
arr = np.arange(10000)
```

```python
for i in range(7):
    key = randint(0, arr[-1] + 1)
    keys.append(key)
    print("Size of array is", arr.size)
    # Linear search
    start = time.time()
    index = linear_search(arr, key)
    time_taken_linear.append(time.time() * 1000 - start * 1000)
    no_of_elem.append(len(arr))
    print("Element", key, "found at", index, end=', ') if index != -1 else print("Element not found.")
    print("Time taken using Linear Search is", time_taken_linear[i], end='\n\n')
    if i == 6:
        break
    arr = np.append(arr, arr[-1] + arr)
```

```python
print(pd.DataFrame({'Time Taken (ms)': time_taken_linear, 'Search Key': keys}))
print("\n\n")
plt.plot(time_taken_linear, no_of_elem)
plt.scatter(time_taken_linear, no_of_elem, label='Linear Search')
plt.title("Performance")
plt.xlabel('Time Taken (ms)')
plt.ylabel('No. of Elements')
plt.legend()
plt.show()
```
