```python
paragraph = 'Heaeas hello eas'
pattern = 'eas'


def search(para, pat):
    flag = 0
    if pat:
        for i in range(len(para)):
            if para[i] == pat[0]:
                if para[i:(i + len(pat))] == pat:
                    print("Pattern found at:" if flag == 0 else '', i, end=' ')
                    flag = 1
    print("Pattern not found" if flag == 0 else '')


search(paragraph, pattern)
```
