# Import Libraries

```python
import time
from random import randint
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
time_taken_binary = []
no_of_elem = []
keys = []
```

```python
def binary_search(a, l, r, x):
    if r >= l:
        mid = l + (r - l) // 2
        if a[mid] == x:
            return mid
        elif a[mid] > x:
            return binary_search(a, l, mid - 1, x)
        else:
            return binary_search(a, mid + 1, r, x)
    else:
        return -1
```

```python
arr = np.arange(10000000)
```

```python
for i in range(7):
    key = randint(0, arr[-1] + 1)
    keys.append(key)
    print("Size of array is", arr.size)
    # binary search
    start = time.time()
    index = binary_search(arr, 0, len(arr) - 1, key)
    time_taken_binary.append(time.time() * 10000 - start * 10000)
    no_of_elem.append(len(arr))
    print("Element", key, "found at", index, end=', ') if index != -1 else print("Element not found.")
    print("Time taken using binary Search is", time_taken_binary[i], end='\n\n')
    if i == 6:
        break
    arr = np.append(arr, arr[-1] + arr)
```

```python
print(pd.DataFrame({'Time Taken (ms)': time_taken_binary, 'Search Key': keys}))
plt.plot(time_taken_binary, no_of_elem)
plt.scatter(time_taken_binary, no_of_elem, label='Binary Search')
plt.xlabel('Time Taken (ms)')
plt.ylabel('No. of Elements')
plt.legend()
plt.show()
```
