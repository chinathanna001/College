```python
def find_min_max(arr, start, end):
    # Base case: If there is only one element in the array
    if start == end:
        return arr[start], arr[start]

    # If there are two elements in the array
    if end - start == 1:
        if arr[start] < arr[end]:
            return arr[start], arr[end]
        else:
            return arr[end], arr[start]

    # Divide the array into two halves
    mid = (start + end) // 2
    min1, max1 = find_min_max(arr, start, mid)
    min2, max2 = find_min_max(arr, mid + 1, end)

    # Compare the minimum and maximum elements of the two halves
    minimum = min(min1, min2)
    maximum = max(max1, max2)

    return minimum, maximum


# Example usage
arr = [5, 9, 3, 1, 8, 6, 2, 7]
start = 0
end = len(arr) - 1
min_element, max_element = find_min_max(arr, start, end)
print("Minimum element:", min_element)
print("Maximum element:", max_element)
```
