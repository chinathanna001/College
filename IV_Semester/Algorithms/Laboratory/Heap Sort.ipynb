```python
import time
import matplotlib.pyplot as plt
import numpy as np

time_taken_insertion = []
time_taken_heap = []
no_of_elem = []


def insertion_sort(a):
    for j in range(1, len(a)):
        key = a[j]
        i = j - 1
        while i >= 0 and a[i] > key:
            a[i + 1] = a[i]
            i -= 1
        a[i + 1] = key


# Heap Sort
def heapify(a, n, i):
    largest = i
    l = 2 * i + 1
    r = 2 * i + 2
    if l < n and a[i] < a[l]:
        largest = l
    if r < n and a[largest] < a[r]:
        largest = r
    if largest != i:
        a[i], a[largest] = a[largest], a[i]
        heapify(a, n, largest)


def heap_sort(a):
    n = len(a)
    for i in range(n, -1, -1):
        heapify(a, n, i)
    for i in range(n - 1, 0, -1):
        a[i], a[0] = a[0], a[i]
        heapify(a, i, 0)


arr = np.random.randint(0, 1000, 1000)

for i in range(4):
    print("Size of array is", arr.size)
    # insertion sort
    x = arr.copy()
    start = time.time()
    insertion_sort(x)
    time_taken_insertion.append(time.time() - start)
    print("Time taken using Insertion sort is", time_taken_insertion[i])
    # heap sort
    y = arr.copy()
    start = time.time()
    heap_sort(y)
    time_taken_heap.append(time.time() - start)
    no_of_elem.append(len(arr))
    print("Time taken using Heap sort is", time_taken_heap[i], end='\n\n')
    if i == 6:
        break
    arr = np.append(arr, arr[-1] + arr)
plt.plot(time_taken_heap, no_of_elem, label='Heap sort')
plt.plot(time_taken_insertion, no_of_elem, label='Insertion sort')
plt.xlabel('Time Taken (s)')
plt.ylabel('No. of Elements')
plt.legend()
plt.show()
```
